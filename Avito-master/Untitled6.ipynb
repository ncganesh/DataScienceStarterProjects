```
def cleantext(text):
  text = "".join(word for word in text.split() if 'http' not in word and not word.startswith ('@') and word != 'RT')
  letters = re.sub("[^a-zA-Z]","",text)
  words = letters.lower().split()
  stop = set(stopwords.words("english"))
  cleword = [w for w in words if not w in stop]
  return cleword
```
